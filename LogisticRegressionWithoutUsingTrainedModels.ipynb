```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
x=np.linspace(-6,6,100)
```

```python
y=sigmoid(x)
```

```python
plt.plot(x, y, label='Logistic (Sigmoid) Function', color='blue')
plt.xlabel('X')
plt.ylabel('Sigmoid(X)')
plt.title('Logistic (Sigmoid) Function')
plt.grid()
plt.legend()
plt.axhline(0.5, color='red', linestyle='--', label='Decision Boundary (y=0.5)')

plt.show()
```
