```python
def petal_length(num):
  if num==1.4:
    return "perfect size"
  elif num>1.4:
    return "Quite bigger"
  elif num<1.4 and num>0:
    return "small size"
  else:
    return "data is invalid"
```

```python
df['PetalLengthCm'].apply(petal_length)
```

```python
df['type_of_petal_length']=df['PetalLengthCm'].apply(petal_length)
```
